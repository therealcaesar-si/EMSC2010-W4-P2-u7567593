## Plotting the Keeling curve
Monthly mean carbon dioxide has been measured at Mauna Loa Observatory, Hawaii, since 1958.  The carbon dioxide data on Mauna Loa constitute the longest record of direct measurements of CO$_2$ in the atmosphere.

This record is referred to as the *Keeling curve* in honour of Charles Keeling who started the monitoring program and ran it until his death in 2005.

```python
import pandas as pd #used to read and clean data
import matplotlib.pyplot as plt #used for data plotting
```

```python
spreadsheet_name = 'co2_mm_mlo.xlsx' #Add the data file into colab

df_co2 = pd.read_excel(spreadsheet_name) #read the co2 data into a dataframe

df_co2.head() #shows us the start of the spreadsheet
```

```python
fig, ax = plt.subplots(figsize=(10, 4)) #make axes (called ax) 10 units wide and 4 units high
ax.plot(df_co2['decimal date'],df_co2['average'],'k') #plot decimal date vs. average co2 on ax as a black ('k') line
ax.set_xlabel('Time (decimal date)') #text for the x-axis label
ax.set_ylabel('Average CO$_2$ (ppm)') #text for the y-axis label
#the "$_2$"" from the above line gave us the subscript 2 for CO2, no need for more installing, its part of matplotlib
ax.set_title('Mauna Loa Monthly Carbon Dioxide') #text for the title
#ax.set_xlim([0,120]) #interval for the x-axis, not needed rn
#ax.set_ylim([30,65]) #interval for the y-axis, not needed rn
ax.minorticks_on() #show the minor ticks
ax.grid() #show gridlines...
```
